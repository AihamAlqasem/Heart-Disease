```python
import pandas as pd
colomns=['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','target']
df=pd.read_csv('processed.cleveland.csv',names=colomns)
df.info()
```

```python
df['ca'] = pd.to_numeric(df['ca'], errors='coerce')# errors هذا يعني اذا لم تستطع تحويل القيم اكتب عليها nan
df['thal'] = pd.to_numeric(df['thal'], errors='coerce')
```

```python
df['target'] = (df['target'] > 0).astype(int)
```

```python
df['target'].value_counts()
```

```python
# df.info()#لطباعة المعلومات عن البيانات
# df.dropna(inplace=True)# حذف صفوف التي تحمل قيم فاضية
# df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x=df.drop('target',axis=1)# في حال كتبت =0 بياخذ الصفوف . ملاحظة في حال كتبت inplace=true عيحذف من الاصلي
y=df['target']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(class_weight='balanced',random_state=42)
model.fit(x_train,y_train)
```

```python
y_rf_pred=model.predict(x_test)
from sklearn.metrics import accuracy_score
pred=accuracy_score(y_test,y_rf_pred)
print(f"الدقة هي :{pred:.2%}")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_rf_pred,y_test))
```
